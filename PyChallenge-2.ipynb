```python
import urllib2
from HTMLParser import HTMLParser
import string
```

```python
page = urllib2.urlopen('http://www.pythonchallenge.com/pc/def/ocr.html')
```

```python
class MyParser(HTMLParser):
    commentlist = []
    
    def handle_comment(self, data):
        self.save_comment(data)
        print "Found %s comment(s)!" % len(self.commentlist)
        
    def save_comment(self, data):
        self.commentlist.append(data)
        return
        
    def get_comments(self):
        return self.commentlist
```

```python
htmldoc = page.read()
```

```python
htmldoc
```

```python
parser = MyParser()
parser.feed(htmldoc)
```

```python
print parser.get_comments()
```

```python
thing = parser.get_comments()[1]
```

```python
lettercounts = {}
for letter in thing:
    if letter in lettercounts.keys():
        lettercounts[letter] += 1
    else:
        lettercounts[letter] = 1        
```

```python
print lettercounts
```

```python
unphrase = [ key for key in lettercounts.keys() if lettercounts[key] > 1 ]
''.join(unphrase)
```

```python
for char in thing:
    if char not in unphrase:
        print char
```
