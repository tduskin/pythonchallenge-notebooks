```python
from bs4 import BeautifulSoup
import urllib2
```

```python
url = 'http://mdlz-rest-portal.starmobileinc.com/api/InstanceManager'
```

```python
password_mgr = urllib2.HTTPPasswordMgrWithDefaultRealm()
top_level_url = url
username = 'admin'
password = 'open123!'
password_mgr.add_password(None, top_level_url, username, password)
handler = urllib2.HTTPBasicAuthHandler(password_mgr)
opener = urllib2.build_opener(handler)
opener.open(url)
urllib2.install_opener(opener)
```

```python
page = urllib2.urlopen(url)
htmldoc = page.read()
```

```python
print htmldoc
```

```python
soup = BeautifulSoup(htmldoc)
```

```python
instancetable = soup.find_all("table")[0]
```

```python
for row in instancetable.find_all("tr")[2:]:
    cells = row.find_all("td")
    
    instance_id = cells[1].string
    instance_ip = cells[2].string
    instance_state = cells[3].string
    
    record = {"id": instance_id, "ip_addr": instance_ip, "state": instance_state}
    print record
```
